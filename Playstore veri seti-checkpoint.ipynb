```python
for col in playstorus.columns:
    col1 = col.replace(' ','')
    playstorus = playstorus.rename(columns={col:col1})
playstorus.columns
```

```python
print("Eksik değerleri içeren sütunlar")
playstorus.isnull().any()
```

```python
print("Veri kümesinde boş değerlere sahip satır sayısı:")
missing_info = (len(playstorus[playstorus.isnull().any(axis=1)]) / len(playstorus) )*100
print(len(playstorus[playstorus.isnull().any(axis=1)]),' adetin ' ,round(missing_info,2) , '%')
```

```python
playstorus.info()
```

```python
playstorus.isnull().sum()
```

```python
playstorus.dropna(subset=["Size","MinimumInstalls","AppName","Installs"],inplace=True)
playstorus.isnull().sum() #veriler temizlendi
```

```python
playstorus["ContentRating"].value_counts()
```

```python
playstorus["ContentRating"]=playstorus["ContentRating"].replace("Unrated","Everyone")
playstorus["ContentRating"]=playstorus["ContentRating"].replace("Everyone 10+","Everyone")
playstorus["ContentRating"]=playstorus["ContentRating"].replace("Adults only 18+","Adults")
```

```python
playstorus.Installs = playstorus.Installs.str.replace(',','')
playstorus.Installs = playstorus.Installs.str.replace('+','')
playstorus.Installs = playstorus.Installs.str.replace('Free','0')
playstorus['Installs'] = pd.to_numeric(playstorus['Installs'])
```

```python
playstorus['PriceRange'] = pd.cut(playstorus['Price'],bins=[0,0.19,9.99,29.99,410],labels=['Free','Low','Mid','High'],include_lowest=True)
playstorus['PriceRange'].value_counts() #sütunları da temizledik
```

```python
playstorus['Type'] = np.where(playstorus['Free'] == True,'Free','Paid')
playstorus.drop(['Free'],inplace=True,axis=1)
```

```python
playstorus.info()
```

```python
playstorus["Type"].value_counts() #ilginç bir şekilde depolama kullanımı daha da artıyor sadness
```

```python
playstorus['RatingType'] = 'NoRatingProvided'
playstorus.loc[(playstorus['RatingCount'] > 0) & (playstorus['RatingCount'] <= 10000.0),'RatingType'] = 'Less than 10K'
playstorus.loc[(playstorus['RatingCount'] > 10000) & (playstorus['RatingCount'] <= 500000.0),'RatingType'] = 'Between 10K and 500K'
playstorus.loc[(playstorus['RatingCount'] > 500000) & (playstorus['RatingCount'] <= 138557570.0),'RatingType'] = 'More than 500K'
playstorus.RatingType.value_counts()
```

```python
playstorus.info()
```

```python
print('Gençlerin ilk 5 seçimi')
adultTopChoices = playstorus.loc[(playstorus.ContentRating == 'Teen')]
adultTopChoices.sort_values(by='MaximumInstalls',ascending=False).head(5)
```

```python
top3cat=playstorus.Category.value_counts().head(3)
top3cat.plot(kind='bar', figsize=(6,6))
plt.title("En iyi 3 kategori")
plt.xlabel("Kategoriler")
plt.ylabel("Oran")
plt.xticks(rotation=0) #biraz da grafik karabiber tuz
```

```python
paid_apps = playstorus[(playstorus.Type=='Paid') & (playstorus.Installs == 5000000) | ((playstorus.Type=='Paid') & (playstorus.Installs == 10000000))]
paid_apps = paid_apps.groupby(['AppName'])['Rating'].max().sort_values(ascending=False)
paid_apps = paid_apps.head(10)
plt.title("En yüksek puanlara ve yüklemelere dayalı En Popüler Ücretli Uygulamalar")

ax = sns.lineplot(x=paid_apps.values,y=paid_apps.index,color='Red');
```

```python
free_apps = playstorus[(playstorus.Type=='Free') & (playstorus.Installs == 5000000) | ((playstorus.Type=='Free') & (playstorus.Installs == 10000000))]
free_apps = free_apps.groupby(['AppName'])['Rating'].max().sort_values(ascending=False)
free_apps = free_apps.head(5)
plt.title("En yüksek derecelendirmelere ve yüklemelere dayalı En İyi Ücretsiz Uygulamalar")

ax = sns.lineplot(x=free_apps.values,y=free_apps.index,color='blue');
```

```python
TeenCatPre= playstorus[(playstorus.ContentRating=='Teen')]
TeenCatPre= TeenCatPre.groupby(['Category'])['MaximumInstalls'].max().sort_values(ascending=False)
TeenCatPre= TeenCatPre.head(10)
pie, ax = plt.subplots(figsize=[10,6])
labels = TeenCatPre.keys()
plt.pie(x=TeenCatPre, autopct="%.1f%%", explode=[0.05]*10, labels=labels, pctdistance=0.5)
plt.title("Gençlerin kategori bazında uygulama yükleme dağılımları", fontsize=14);
```
